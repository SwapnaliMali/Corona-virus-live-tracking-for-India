# Corona virus live tracking for India

```python
import pandas as pd
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
#removing the srno. column and the 36 th row since it contains the total of above only.
df=df.iloc[:-1,1:]# remove last row ,  remove 1st column.
df
```

```python
df.tail(5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
total_states=np.arange(len(df['state_name']))# we first take le of state names i.e 36. the convert it to numpy array, store it in total_states variable.It means 0=andaman & nicobar etc.later we can convert it back to names 
total_states
```

```python
df['active']#active cases
```

```python
max(df['positive'])
```

# total positive cases based on states in India

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')    #r,g,b
plt.barh(total_states,df['positive'],align='center',alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])#y axis- staate names.
plt.xlim(1,max(df['positive'])+100)# xlim should be equal to max positive cases + 10.
plt.xlabel('positive number of cases')
plt.title('corona virus cases')
plt.show()
```

# Active Number of Cases per State

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')    #r,g,b
plt.barh(total_states,df['active'],align='center',alpha=0.5,color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])#y axis- staate names.
plt.xlim(1,max(df['active'])+100)# xlim should be equal to max positive cases + 10.
plt.xlabel('positive number of cases')
plt.title('corona virus cases')
plt.show()
```

# Death cases per state

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')    #r,g,b
plt.barh(total_states,df['death'],align='center',alpha=0.5,color=(0,1,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])#y axis- staate names.
plt.xlim(1,max(df['death'])+100)# xlim should be equal to max positive cases + 10.
plt.xlabel('death number of cases')
plt.title('corona virus cases')
plt.show()
```
